```python
import pandas as pd
import numpy as np
from nltk.tokenize import TreebankWordTokenizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer, TfidfTransformer
from sklearn.model_selection import train_test_split

t = pd.read_csv('ExtractedTweets.csv')
h = pd.read_csv('TwitterHandles.csv')

train_data, test_data = train_test_split(t[['Party', 'Tweet']], test_size=0.5)

vect = CountVectorizer()
tokenizer = TreebankWordTokenizer()
transformer = TfidfTransformer(smooth_idf=False)

#vect.set_params(tokenizer=tokenizer.tokenize)
vect.set_params(token_pattern='\\b[A-Za-z]+\\b')
vect.set_params(stop_words='english')
vect.set_params(ngram_range=(1, 2))
vect.set_params(min_df=2)

train_data_counts = vect.fit_transform(train_data['Tweet'])
train_weighted_counts = transformer.fit_transform(train_data_counts)
vect.get_feature_names()
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(train_data_counts, train_data['Party'])
```

```python
test_data_counts = vect.transform(test_data['Tweet'])
test_weighted_counts = transformer.transform(test_data_counts)

predictions = classifier.predict(test_weighted_counts)

correct_count = 0
for actual, predicted in zip(test_data['Party'], predictions):
    if actual == predicted:
        correct_count += 1
correct_count / len(predictions)
```
